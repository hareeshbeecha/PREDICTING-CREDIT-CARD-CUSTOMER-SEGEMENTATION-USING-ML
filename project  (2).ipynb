## Grouping

#### Calculating the average customer age for each gender

```python
df.groupby('Gender')['Customer_Age'].mean()
```

#### Calculating the total transaction amount for each income category

```python
df.groupby('Income_Category')['Total_Trans_Amt'].sum()
```

 #### Grouping by Gender: Average Credit Limit and Total Transaction Amount

```python
gender_group = df.groupby('Gender').agg({'Credit_Limit': 'mean', 'Total_Trans_Amt': 'sum'}).reset_index()
print("Grouping by Gender - Average Credit Limit and Total Transaction Amount:")
print(gender_group)
```

### Visualizations

### Histogram for customer age distribution

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Customer_Age'], bins=20, color='skyblue', edgecolor='black')
plt.title("Customer Age Distribution")
plt.xlabel("Customer Age")
plt.ylabel("Count")
plt.show()
```

### Box Plot for age distribution by gender

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Customer_Age', data=df)
plt.title("Customer Age Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Customer Age")
plt.show()
```

### Scatter Plot: Customer Age Distribution by Attrition

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Attrition_Flag', y='Customer_Age', data=df)
plt.title("Customer Age Distribution by Attrition")
plt.xlabel("Attrition Flag")
plt.ylabel("Customer Age")
plt.show()
```

### Pie chart for gender distribution

```python
plt.figure(figsize=(6, 6))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

### Bar Graph for Income Category Distribution by Gender

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Income_Category', y='Customer_Age', hue='Gender', data=df)
plt.title("Income Category Distribution by Gender")
plt.xlabel("Income Category")
plt.ylabel("Customer Age")
plt.xticks(rotation=45)
plt.legend(title="Gender", loc="upper right")
plt.show()
```

### Barchart for Average Transaction Amount by Education Level

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Education_Level', y='Total_Trans_Amt', data=df)
plt.title("Average Transaction Amount by Education Level")
plt.xticks(rotation=45)
plt.ylabel("Average Transaction Amount")
plt.show()
```

### Graph for attrition by education level

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Education_Level', hue='Attrition_Flag', data=df)
plt.title("Attrition by Education Level")
plt.xticks(rotation=45)
plt.show()
```

### Bar Graph for average utilization ratio by education level

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Education_Level', y='Avg_Utilization_Ratio', data=df)
plt.title("Average Utilization Ratio by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Average Utilization Ratio")
plt.xticks(rotation=45)
plt.show()
```

### Distribution of Education Level

```python
Education = df['Education_Level'].value_counts().sort_values()
plt.figure(figsize=(10,5))  
plt.bar(Education.index,Education)
plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Distribution of Education Level')
plt.show()
```
